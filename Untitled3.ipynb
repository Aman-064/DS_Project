```python
# 1. Fill missing values
df['Discount Name'] = df['Discount Name'].fillna('No Discount')
```

```python
# 2. Fix Formatting
df['Net Amount'] = pd.to_numeric(df['Net Amount'].astype(str).str.replace(',', ''), errors='coerce')
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'], errors='coerce', dayfirst=True)
```

```python
# 3. Clean up invalid data & drop duplicates
df = df[df['Net Amount'] >= 0]
df = df.drop_duplicates()
df
```

```python
# Optionally, save your cleaned data
df.to_csv('cleaned_ecommerce_df.csv', index=False)
```

```python
print("Data cleaning completed successfully!")
```

```python
# Look at overall summary statistics (averages, min, max, etc.)
summary_stats = df[['Gross Amount', 'Net Amount', 'Discount Amount (INR)']].describe()
print("Overall Sales Statistics:\n", summary_stats)
```

```python
# Analyze trends: What is the average Net Amount spent per Product Category?
avg_net_by_category = df.groupby('Product Category')['Net Amount'].mean().sort_values(ascending=False)
print("\nAverage Net Amount by Category:\n", avg_net_by_category)
```

```python
import matplotlib.pyplot as plt
```

```python
# 1. Bar Graph: Number of Purchases by Product Category
plt.figure(figsize=(10, 6))
category_counts = df['Product Category'].value_counts()
plt.bar(category_counts.index, category_counts.values, color='teal')
plt.title('Number of Purchases by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 2. Histogram: Distribution of Net Amount
plt.figure(figsize=(10, 6))
plt.hist(df['Net Amount'], bins=50, color='royalblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Net Purchase Amount')
plt.xlabel('Net Amount (INR)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# 3. Bar Graph: Average Net Amount by Gender
plt.figure(figsize=(8, 5))
gender_avg = df.groupby('Gender')['Net Amount'].mean()
plt.bar(gender_avg.index, gender_avg.values, color=['coral', 'mediumseagreen', 'mediumpurple'])
plt.title('Average Net Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Net Amount (INR)')
plt.tight_layout()
plt.show()
```
